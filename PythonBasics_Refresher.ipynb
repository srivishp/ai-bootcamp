# range() function
range(start, stop, step)

#### start (Optional): 
An integer specifying the starting point of the sequence. If omitted, it defaults to 0.
#### stop (Required): 
An integer specifying the upper limit of the sequence. The sequence will stop before this number.
#### step (Optional): 
An integer specifying the increment between numbers in the sequence. If omitted, it defaults to 1. A negative step value can be used to count backward.

```python
for x in range(1,10):
    print (x, end = " ")
```

```python
list(range(0,48, 4))
```

```python
for x in range(0,128, 4):
    print (x, end = " ")
```

```python
list(range(15))
```
